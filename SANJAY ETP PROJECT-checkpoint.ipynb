```python
from tkinter import *
from tkinter import messagebox
import time

root = Tk()
root.title('Tic-Tac-Toe_game')
#players
Players = ['O', 'X']

#first players turn
turn = '1st'

#initially the count is 0
count = 0
# message  = true is the variable taken for button click
message = True

def disable_all_buttons():  #if winner is declared it will disable all the buttons and displays the winner
    for k in Buttons:
        k.config(state=DISABLED)

def checkifwon():
    global winner, count, messa
    winner = False #means there is no winner for now
    win_index = [] #empy list later indexes will be stored in it
    for x in Players: #this forloop runs for both the players 
        # for i in range(9): for loop Generate numbers between 0 to 9 for these 9 buttons we have 
        #to check if there is any winng pattern in rows
        for i in range(9):
            if i == 0 or i == 3 or i == 6: #starts from 1st button row 1 row2  row 3
                if Buttons[i]['text'] == x and Buttons[i+1]['text'] == x and Buttons[i+2]['text'] == x: #for player 1
                    win_index = [i, i+1, i+2] #winning pattern
                    winner = True #displays the winner
                    
            if i == 0 or i == 1 or i == 2:#starts from 1st button column 1 column2  column 3
                if Buttons[i]['text'] == x and Buttons[i+3]['text']==x and Buttons[i+6]['text'] == x: #player 
                    #described in for loop 
                    win_index = [i, i + 3, i + 6] #winning pattern
                    winner = True   #displays the winner
                    
                    #1st diagonal winning pattern that starts from 0
            if i == 0:
                if Buttons[i]['text'] == x and Buttons[i+4]['text'] ==x and Buttons[i+8]['text'] == x:
                    win_index = [i, i + 4, i + 8]
                    winner = True
            if i == 2: #2nd diagonal winning pattern starts from 2
                if Buttons[i]['text'] == x and Buttons[i+2]['text'] ==x and Buttons[i+4]['text'] == x:
                    win_index = [i, i + 2, i + 4]
                    winner = True
                    
                    #if there is winner
        if winner == True:
            for k in win_index:
                Buttons[k].config(highlightbackground='Red')
            b = f'Congratulations {x}, you won!' # creating f string and display the message
    #root.update()to allow tasks such as refreshing a GUI to happen immediately
            root.update()
    #time.sleep(1)used to add delay in the execution of a program.
            time.sleep(1)
            messagebox.showinfo("Tic-Tac-Toe", b)
            next_round = messagebox.askquestion("Tic-Tac-Toe", " Want to Play Another Round? ")
            if next_round == 'yes':
                count = 0
                for k in Buttons:
                    k['text'] = ' '
                    k.config(highlightbackground='Yellow')
            elif next_round == 'no':
                disable_all_buttons()
                exit()
            break
        elif winner == False and count == 9: # if all the buttons get clicked and there is no winner 
            root.update()
            time.sleep(1)
            #display message 
            next_round = messagebox.askquestion("Tic-Tac-Toe", 'It is a Tie No one wins \n Want to play another Round?')
            #ask whether to go to next round
            
            if next_round == 'yes':
                for k in Buttons:
                    k['text'] = ' '
                break
            elif next_round == 'no':
                disable_all_buttons()
                root.quit()
                
#click function
#b is button widget
def click(b):
    #we need to call turn ,count to global button to get responses
    global turn, count
    b = b.widget
    txt = b['text']
    if txt == ' ' and turn == '1st':
        #list subscription method to display names of the players on the widget
        b['text'] = Players[0]
        count += 1
        turn = '2nd'
        checkifwon()
        
        #if the click is made by 2nd player in that case the 1st requirement is that text must be 
        #empty and 2nd thing is that it must be the turn of 2nd player
    elif txt == ' ' and turn == '2nd':
        b['text'] = Players[1]
        count += 1
        turn = '1st'
        checkifwon()
        
#creation of buttons using f string 9 buttons will be created we will get list of characters from b1,b2,..........,b9
buttons = [f'b{i}' for i in range(1, 10)]
Buttons = []
for k in buttons:
    globals()[k] = Button(root, text=' ', width=20, height=10, highlightbackground='yellow')
    #appending buttions into button list
    Buttons.append(globals()[k])
    #whenever we click on button some action must happen
    globals()[k].bind('<Button-1>', click)
    
    
#packing the buttons into rows and columns
row = 0
column = 0

#for loop for 9 buttons
for i in range(9):
    Buttons[i].grid(row=row, column=column)
    #with each iteration there should be increment in the column
    column += 1
    #whenever i reaches 5 row must increase by 1 and when index goes to 2 column must reach to 0
    if i == 2 or i == 5:
        row += 1
        column = 0
        #root.mainloop() executes what we wish to execute in an application
root.mainloop()
```
